```python
import jqdatasdk
username = input('Please input your username (phone number):')
password = input('Input your pass word:')
jqdatasdk.auth(username, password)
query_count = jqdatasdk.get_query_count()
print('spare / total: {} / {}'.format(query_count['spare'], query_count['total']) )
```

```python
ori_code = ['sz300800', 'sz300805']
target = jqdatasdk.normalize_code(ori_code) # 'sz300805'
default_fields = ['open','low','high','close','volume','factor']
fields_info = ['']
```

```python
d = jqdatasdk.get_price(target, start_date="2019-11-21", end_date="2019-12-01", fields=default_fields, fq='post')
```

```python
d
```

```python
out = []
for k, ori_k in zip(target, ori_code):
    d_k = d[d['code'] == k]
    d_k.loc[:,'code'] = ori_k
    d_k.rename(columns={'code':'symbol'}, inplace=True)
    out.append(d_k)
```

```python
out[1]
```

```python
d = jqdatasdk.get_price(target, start_date="2019-11-21", end_date="2019-12-01", fields=default_fields, fq='post')
d = d.reset_index()
d.columns = ['date', 'open', 'low', 'high', 'close', 'volume', 'factor']
d['symbol'] = ori_code  # 增加symbol
```

```python
d.to_csv(ori_code + ".csv", index=False)
```
